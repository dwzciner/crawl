```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import jieba  
import requests
from wordcloud import WordCloud
%matplotlib inline 
sns.set(style='darkgrid')  
```

```python
df = pd.read_csv(r"C:\Users\wang'zi'cheng\Desktop\city_2_analysis.csv")
```

```python
df.head()
```

```python
scatter = go.Scattermapbox(lat=df['lat']       #设置纬度
                           ,lon=df['lon']      #设置经度
                             ,name='城市',
                           hover_data='count'
                          )                           #使用Scattermapbox函数画散点图
fig = go.Figure(scatter)       #将散点图导入画布
fig.update_layout(mapbox_style='stamen-terrain')      #将地图设置为画布
#可以使用的免费地图："open-street-map", "carto-positron", "carto-darkmatter", "stamen-terrain", "stamen-toner" or "stamen-watercolor"
fig.show()
```

```python
import plotly.express as px
fig = px.scatter_geo(df,
                     hover_name="城市", size="count",
                     projection="natural earth")
fig.show()
```

```python
import plotly.express as px 
#在地图上绘制密度图

fig = px.density_mapbox(df
                        , lat='lat'       #设置纬度
                        ,lon='lon'        #设置经度
                        , z=df['count']/10000                #设置密度
                        , radius=20,
                        center=dict(lat=40, lon=118)            #设置地图的中心点
                        , zoom=2                               #设置地图的缩放等级0-20
                        ,mapbox_style="stamen-terrain"
                       ,color_continuous_scale='inferno'
                       )        #设置地图的底图

fig.show()
```
